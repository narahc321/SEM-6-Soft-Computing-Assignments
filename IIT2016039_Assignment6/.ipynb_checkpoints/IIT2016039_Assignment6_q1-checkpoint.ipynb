```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
import cv2

images =[]
for i in range(1,41,1):
    s = 'data/training/'+str(i)+'.jpg'
    img = cv2.cvtColor(cv2.imread(s), cv2.COLOR_RGB2GRAY)
    img = np.array(img)
    img = img.reshape([img.shape[0]*img.shape[1]])
    images.append(img)
imagess = np.array(images)

length = 1280
width = 720
```

```python
avg = np.average(imagess,axis=0)
imgplot = plt.imshow(avg.reshape([length,width]),cmap='gray')
images = imagess - avg
cov = np.cov(images)
eig_vals, eig_vecs, = np.linalg.eig(cov)
eigs = [(eig_vals[i], i) for i in range(len(eig_vals))]
eigs.sort(reverse=True)
eig_vecs_sort = np.array([eig_vecs[eigs[i][1]] for i in range(len(eig_vals))])
k = 10
feature_vectors = eig_vecs_sort[:k, :]
eigen_faces = feature_vectors.dot(images)
W = images.dot(eigen_faces.T)
```

```python
for i in range(1,9,1):
    test = cv2.cvtColor(cv2.imread('data/testing/'+str(i)+'.jpg'), cv2.COLOR_RGB2GRAY)
    test_vector = np.array(test).flatten()
    mean_test = np.subtract(test_vector,avg)
    mean_test = mean_test.reshape([-1, 1])
    W_test = eigen_faces.dot(mean_test)
    diff  = W - W_test.T
    index = np.argmin(np.linalg.norm(diff, axis=1))
    img =mean_test.reshape([length,width])
    plt.subplot(1,2,1)
    plt.imshow(test, cmap='gray')
    plt.subplot(1,2,2)
    plt.imshow(cv2.cvtColor(cv2.imread('data/training/'+str(index)+'.jpg'), cv2.COLOR_RGB2GRAY), cmap='gray')
    plt.show()
```

```python
accuracies = []
x = []
for k in range(1,31):
    feature_vectors = eig_vecs_sort[:k, :]
    eigen_faces = feature_vectors.dot(images)
    W = images.dot(eigen_faces.T)
    total = 0
    correct = 0
    for i in range(1,9,1):
        test = cv2.cvtColor(cv2.imread('data/testing/'+str(i)+'.jpg'), cv2.COLOR_RGB2GRAY)
        test_vector = np.array(test).flatten()
        mean_test = test_vector - avg
        mean_test = mean_test.reshape([-1, 1])    
        W_test = eigen_faces.dot(mean_test)
        diff  = W - W_test.T
        index = np.argmin(np.linalg.norm(diff, axis=1))
        img =mean_test.reshape([length,width])
        total += 1
        if i==(index//5)+1:
            correct += 1
    x.append(k)
    accuracies.append(correct/total)
plt.plot(x,accuracies,'-yo')
plt.grid()
plt.show()
```
