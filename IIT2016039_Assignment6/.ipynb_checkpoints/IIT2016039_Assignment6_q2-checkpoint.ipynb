```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
import cv2

images =[]
for i in range(1,41,1):
    s = 'data/training/'+str(i)+'.jpg'
    img = cv2.cvtColor(cv2.imread(s), cv2.COLOR_RGB2GRAY)
    img = np.array(img)
    img = img.reshape([img.shape[0]*img.shape[1]])
    images.append(img)
imagess = np.array(images)

length = 1280
width = 720
```

```python
avg = np.average(imagess,axis=0)
imgplot = plt.imshow(avg.reshape([length,width]),cmap="gray")
images = imagess - avg
cov = np.cov(images)
eig_vals, eig_vecs, = np.linalg.eig(cov)
eigs = [(eig_vals[i], i) for i in range(len(eig_vals))]
eigs.sort(reverse=True)
eig_vecs_sort = np.array([eig_vecs[eigs[i][1]] for i in range(len(eig_vals))])
feature_vectors = eig_vecs_sort[0:20, :]
eigen_faces = feature_vectors.dot(images)
W = images.dot(eigen_faces.T)
```

```python
sw = np.zeros((20, 20))
for i in range(0,40,5):
    csw = W[i:i+5,:]
    csw = csw - np.mean(csw,axis=0)
    sw += (csw.T).dot(csw)

sb = np.zeros((20, 20))
mean = np.mean(W.T, axis=1).reshape([-1, 1])
for i in range(0,40,5):
    csb = W[i:i+5,:]
    csb_mean = np.mean(csb,axis=0).reshape([-1, 1])
    diff = csb_mean - mean
    sb += diff.dot(csb_mean.T)

J = np.linalg.inv(sw).dot(sb)
j_eig_vals, j_eig_vecs = np.linalg.eig(J)
j_eigs = [(j_eig_vals[i], i) for i in range(len(j_eig_vals))]
j_eigs.sort(reverse=True)
j_eig_vecs_sort = np.array([j_eig_vecs[j_eigs[i][1]] for i in range(len(j_eig_vals))])
new_eigens = j_eig_vecs_sort[:15,:]
FF = (new_eigens).dot(W.T)
```

```python
for i in range(1,9,1):
    test_face = cv2.cvtColor(cv2.imread('data/testing/'+str(i)+'.jpg'), cv2.COLOR_RGB2GRAY)
    test_face_vector = np.array(test_face, dtype='float64').flatten()
    test_vector = test_face_vector - avg
    test_vector = test_vector.reshape([-1, 1])
    PEF = test_vector.T.dot(eigen_faces.T)
    Projected_Fisher_Test_Img = new_eigens.dot(PEF.T)
    index = np.argmin(np.linalg.norm(FF.T - Projected_Fisher_Test_Img.T,axis=1))
    plt.subplot(1,2,1)
    plt.imshow(test_face, cmap='gray')
    plt.subplot(1,2,2)
    plt.imshow(cv2.cvtColor(cv2.imread('data/training/'+str(index)+'.jpg'), cv2.COLOR_RGB2GRAY), cmap='gray')
    plt.show()
```
